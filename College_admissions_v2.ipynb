# 5. Machine Learning

At the very start, we need to check the correlations of the various features we have in our dataset.

In order to align our data better, let us remove all the unnecessary columns

```python
df.head()
```

```python
df.drop('Serial No.',axis=1,inplace=True)
df.drop('Scaled CGPA',axis=1,inplace=True)
```

Now, we need to separate the target label and inputs from our dataframe into X and Y variables. Let us do that.

```python
Y=df['Chance of Admit '] #Target
X=df.drop('Chance of Admit ',axis=1) #Input
```

 As we can see from our dataframe, the various inputs have extremely different ranges of values. For example, GRE scores are triple digit values while research feature is a binary value. Hence, in order to give equal weights to each feature, it is important to apply standard scale transformation. Let us do that.
 

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
```

Great ! So we have scaled our data. All we gotta do now is to split our input data into train and test data. 

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_scaled,Y,test_size=0.2,shuffle=True,random_state=0)
```

### Linear Regression

Let's start off with a simple linear regression and check how close are we to the actual results


```python
from sklearn.linear_model import LinearRegression
```

```python
lin_reg=LinearRegression()
lin_reg.fit(X_train,y_train)
```

```python
lin_reg.score(X_train,y_train)
```

```python
y_pred_lin=lin_reg.predict(X_test)
lin_reg.score(X_test,y_test)
```

The linear regression scores are extremely average. We should be able to do better than a simple linear regression. Let's check the mean squared error of our predictions.

```python
from sklearn.metrics import mean_squared_error
mse_lin_reg=np.sqrt(mean_squared_error(y_pred_lin,y_test))
print('Root mean squared error of linear regression : {0:.2f}'.format(mse_lin_reg))
```

The MSE is decent with 0.06 . However, we shall aim to do better than this using other regression models.

Let us check how the predictions look on a plot.

```python
lin_reg_df=pd.DataFrame()
lin_reg_df['Target']=y_test
lin_reg_df['Predictions']=y_pred_lin

sns.lmplot('Target','Predictions',data=lin_reg_df,line_kws={'color':'red'},height=5,aspect=2)
plt.title('Linear Regression results \n  RMSE={0:.2f}'.format(mse_lin_reg),size=15)
```

As we can see see, the predictions and targets are somewhat following a linear line which is ideal. We need to try to reduce the scattering of these points about the red line. The closer these points are to the straight line, the higher accuracy we have in our results.
