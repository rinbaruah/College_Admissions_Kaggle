# 4. Data Visualisation

In this section, we will be trying to plot various values of our features that could give us an idea of the important features for admission predictions.

## GRE 

Let us check how the GRE scores influence the Admission chances.

```python

g=sns.lmplot('GRE Score','Chance of Admit ',data=df,order=1,line_kws={'color':'red','linewidth':2.5},
           height=5,aspect=2,scatter_kws={'s':50,'alpha':0.4})
plt.title('GRE score Vs Admission chances',size=25)

plt.axhline(0.8,color='black',alpha=0.2)
plt.axvline(324,color='black',alpha=0.2)
plt.xticks(np.arange(280,365,5))
```

As we can see, there is a roughly linear relation between GRE scores and Admission chances. This is expected since students with high GRE scores tend to get their applications accepted into multiple universities of their choices which increases their chances of an admit.

As we can see from the horizontal and vertical lines here, the regression fitted red line suggests that a score of 324 gives you a 80% chance of getting your preferred admission.


Let us check how the GRE scores are distributed in total and try to find the median and mean of the scores.

```python
plt.figure(figsize=(10,8))
g=sns.distplot(df['GRE Score'],label='GRE scores')
plt.title('GRE Score distribution \n Median:{0:.1f} \n Mean:{1:.1f}'.format(df['GRE Score'].median(),
                                                                            df['GRE Score'].mean()),size=25)
l1=plt.axvline(df['GRE Score'].median(),color='black',label='Median score')
plt.legend()
```

As we can see, the mean and median are almost the same. This means we have a normal distribution of GRE scores which is always preferred while dealing with ML regression later on.

## TOEFL Score

This score is an evaluation of the command over the English language which is necessary for English speaking nations such as United States, Canada, UK, Australia, New Zealand and a few more.

Let's see how the scores are first distributed. 

```python
plt.figure(figsize=(10,8))
g=sns.distplot(df['TOEFL Score'],label='TOEFL scores',color='green')
plt.title('TOEFL Score distribution \n Median:{0:.1f} \n Mean:{1:.1f}'.format(df['TOEFL Score'].median(),
                                                                            df['TOEFL Score'].mean()),size=25)
l1=plt.axvline(df['TOEFL Score'].median(),color='black',label='Median score')
plt.legend()
```

Great ! So, the TOEFL scores are normally distributed too considering that the mean and median are nearly identical.

Let us now check how the TOEFL scores influence the chances of admissions.

```python
sns.lmplot('TOEFL Score','Chance of Admit ',data=df,order=1,line_kws={'color':'red','linewidth':2.5},
           height=5,aspect=2,scatter_kws={'s':50,'alpha':0.4})
plt.title('TOEFL score Vs Admission chances',size=25)

plt.axhline(0.8,color='black',alpha=0.2)
plt.axvline(111,color='black',alpha=0.2)
```

As we can see, the TOEFL scores also have a linear correlation with admission chances. It is seen that a score of about 111 gives us a roughly 80% chance of admissions.

## University Rating

Next, let us see the different ratings of the universities. A higher rating would definitely be a preferred destination for many. 

Let us see how the different universities based on ratings stack up.

```python

sns.catplot('University Rating',data=df,kind='count',height=5,aspect=2)
plt.title('Number of universities of each rating',size=23)
```

From the above barplots, we can see that most universities have a rating of 3 followed by 2 and 4. Usually, the extremities are always least. Hence, number of universities with ratings 1 and 5 are low.

Our intuition says that higher rated universities would have higher GRE and TOEFL score cutoffs. Let us test this theory of ours.

```python
df_r1=df[df['University Rating']==1]
df_r2=df[df['University Rating']==2]
df_r3=df[df['University Rating']==3]
df_r4=df[df['University Rating']==4]
df_r5=df[df['University Rating']==5]
```

```python
fig2=plt.figure(figsize=(10,8))
ax2=fig2.add_subplot(1,1,1)

r1=ax2.scatter(df_r1['GRE Score'],df_r1['TOEFL Score'],color='green',label='Rating 1')
r2=ax2.scatter(df_r2['GRE Score'],df_r2['TOEFL Score'],color='orange',label='Rating 2',)
r3=ax2.scatter(df_r3['GRE Score'],df_r3['TOEFL Score'],color='cyan',label='Rating 3',)
r4=ax2.scatter(df_r4['GRE Score'],df_r4['TOEFL Score'],color='blue',label='Rating 4',)
r5=ax2.scatter(df_r5['GRE Score'],df_r5['TOEFL Score'],color='red',label='Rating 5',)
plt.legend()
plt.title('University rating wise score requirements',size=25)
plt.xlabel('GRE Score')
plt.ylabel('TOEFL Score')
```

As we can see, the red dots in the plot are the 5 star rated universities. As expected, students with high GRE Scores and TOEFL scores have mostly applied for the high rated unversities. Most of the applications are dominated by rating 3 universities in the cyan colors. These are those applicants who have scored between 300-320 in GRE and 100-110 in TOEFL. Scores higher than those have bee found in the higher rated university applications.

## CGPA

Now, we shall evaluate how the CGPA scores influence chances of admit. CGPA is the cumulative grade point score of a student in his undergraduate/post graduate course.

Let us draw a boxplot and a distplot to understand the distribution of the CGPA scores of all the entries.

```python
fig3=plt.figure(figsize=(20,8))
ax3=fig3.add_subplot(121)
ax4=fig3.add_subplot(122)

sns.boxplot(df['CGPA'],orient='v',ax=ax3)
sns.distplot(df['CGPA'],ax=ax4,label='CGPA',color='green')
ax4.axvline(df['CGPA'].median(),color='black',label='Median CGPA')
ax4.set_title('CGPA Distribution \n Median: {}'.format(df['CGPA'].median()),size=25)
plt.legend()
```

Let us see which CGPAs are more likely to get an admit. 

```python
sns.lmplot('CGPA','Chance of Admit ',data=df,line_kws={'color':'red','linewidth':2.5},scatter_kws={'alpha':0.3},height=6,aspect=2)
plt.title('CGPA Vs Chance of admit',size=25)
```

We see that for CGPAs lower than 8.5, the variance in the scattters are high and hence, are not following a linear correlation. However, for CGPAs higher than 8.5, there is a strong linear correlation. Hence, higher CGPAs (above 8.5) are likely (greater than 70% chance) to get an admission in their preferred university.

## LOR and SOP

These are letter of recommendations and statement of purpose which are mandatory for a student application. Statement of purpose generally checks the student's research and why should the university admit a particular student along with the various research outputs that the student can provide with the  limited time they get in the university.

LORs on the other hand serve as an accredition by the student's faculties or peer researchers whose words could help influence the student to get an admit.

Let us now check how these factors play a role in admissions.

```python
fig4=plt.figure(figsize=(14,5))
ax5=fig4.add_subplot(121)
ax6=fig4.add_subplot(122)
g=sns.regplot('SOP','Chance of Admit ',data=df,ax=ax5)
ax5.set_title('SOP  Vs Chance of admit',size=20)
h=sns.regplot('LOR ','Chance of Admit ',data=df,ax=ax6)
ax6.set_title('LOR Vs Chance of Admit',size=20)
```

As we can see, with higher value of SOPs and LORs, the chances of admittance also increase. Hence, high LORs and SOPs are definitely beneficial.


## GRE scores Vs TOEFL scores Vs CGPA

Let us now try to check if there are any common relations amongst GRE, TOEFL scores and CGPA. For this, we shall use scatter plots.

```python
df['Scaled CGPA']=(df['CGPA']-df['CGPA'].min())/(df['CGPA'].max()-df['CGPA'].min())
```

```python
sns.set()
df_high=df[df['Chance of Admit ']>0.7]
df_low=df[df['Chance of Admit ']<0.71]
fig5=plt.figure(figsize=(10,8))
ax7=fig5.add_subplot(1,1,1)
ax7.scatter(df_high['GRE Score'],df_high['TOEFL Score'],alpha=0.5,s=df_high['Scaled CGPA']*200,color='seagreen',label='Chance>70%')
ax7.scatter(df_low['GRE Score'],df_low['TOEFL Score'],alpha=0.5,s=df_low['Scaled CGPA']*200,color='red',label='Chance<70%')
plt.legend(fontsize=15)
plt.title('GRE, TOEFL, CGPA relation \n \n Marker size= CGPA',size=25)
plt.xlabel('GRE Score',size=15)
plt.ylabel('TOEFL Score',size=15)
```

From the above plot, we have made an attempt to check the relation between TOEFL score, GRE score and CGPA value. The **CGPA value is given by the size of the markers on the scatterplot** . The bigger the marker, the higher the CGPA of the applicant.

As it can be seen, students with high GRE scores and TOEFL scores tend to have a high CGPA as well. The green markers also show that the student has a greater than 70% chance of making it to the preferred university.

# 5. Machine Learning

At the very start, we need to check the correlations of the various features we have in our dataset.

In order to align our data better, let us remove all the unnecessary columns

```python
df.head()
```

Now, we need to separate the target label and inputs from our dataframe into X and Y variables. Let us do that.

```python
Y=df['Chance of Admit '] #Target
X=df.drop('Chance of Admit ',axis=1) #Input
```

 As we can see from our dataframe, the various inputs have extremely different ranges of values. For example, GRE scores are triple digit values while research feature is a binary value. Hence, in order to give equal weights to each feature, it is important to apply standard scale transformation. Let us do that.
 

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
```

Great ! So we have scaled our data. All we gotta do now is to split our input data into train and test data. 

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_scaled,Y,test_size=0.2,shuffle=True,random_state=0)
```

### Linear Regression

Let's start off with a simple linear regression and check how close are we to the actual results


```python
from sklearn.linear_model import LinearRegression
```

```python
lin_reg=LinearRegression()
lin_reg.fit(X_train,y_train)
```

```python
lin_reg.score(X_train,y_train)
```

```python
y_pred_lin=lin_reg.predict(X_test)
lin_reg.score(X_test,y_test)
```

The linear regression scores are extremely average. We should be able to do better than a simple linear regression. Let's check the mean squared error of our predictions.

```python
from sklearn.metrics import mean_squared_error
mse_lin_reg=np.sqrt(mean_squared_error(y_pred_lin,y_test))
print('Root mean squared error of linear regression : {0:.2f}'.format(mse_lin_reg))
```

The MSE is decent with 0.06 . However, we shall aim to do better than this using other regression models.

Let us check how the predictions look on a plot.

```python
lin_reg_df=pd.DataFrame()
lin_reg_df['Target']=y_test
lin_reg_df['Predictions']=y_pred_lin
```


### Ridge Regression

Let us use a Ridge regression technique which is similar to simple linear regression albeit, it uses a penalty technique that penalises wrong predictions based on the residual. Lasso regression utilises a L2 regularisation which is squared in nature.

In our case, we have utilised a Ridge regression built in with cross validation. The cross validation technique prevents the overfitting issue.

```python
from sklearn.linear_model import RidgeCV
```

```python
Rid_reg=RidgeCV(cv=10)
Rid_reg.fit(X_train,y_train.values)
```

```python
Rid_reg.score(X_train,y_train)
```

```python
y_pred_rid=Rid_reg.predict(X_test)
Rid_reg.score(X_test,y_test)
```

As we can see, the scores are quite similar to simple linear regression. Hence, even with regularisation and cross validation, we could not improve upon the score much. It seems we need to think past the linear models for getting higher accuracies.

```python
mse_rid_reg=np.sqrt(mean_squared_error(y_pred_rid,y_test))
print('Root mean squared error of Ridge regression : {0:.2f}'.format(mse_lin_reg))
```

```python
Rid_reg_df=pd.DataFrame()
Rid_reg_df['Target']=y_test
Rid_reg_df['Predictions']=y_pred_rid
```

```python
fig6=plt.figure(figsize=(15,5))
ax8=fig6.add_subplot(121)
sns.regplot('Target','Predictions',data=lin_reg_df,line_kws={'color':'red'},ax=ax8)
ax9=fig6.add_subplot(122)
sns.regplot('Target','Predictions',data=Rid_reg_df,line_kws={'color':'red'},ax=ax9,scatter_kws={'color':'green'})

ax8.set_title('Linear Regression results \n  RMSE={0:.3f}'.format(mse_lin_reg),size=15)
ax9.set_title('Ridge Regression results \n  RMSE={0:.3f}'.format(mse_rid_reg),size=15)
```

As we can see see, the predictions and targets are somewhat following a linear line which is ideal. We need to try to reduce the scattering of these points about the red line. The closer these points are to the straight line, the higher accuracy we have in our results.

Also, both simple and ridge regressions have given us nearly identical results. Hence, it seems that the linear model cannot get past a RMSE of 0.06 .

Let us look at some of the regressions that could yield better results.

Maybe tree based ensembles such as Random Forest regressor ? Let's check it out !

### Random Forest 

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
```

```python
rfr=RandomForestRegressor(random_state=0)
param_grid={'n_estimators':[5,7,9,10], 'max_depth':[5,7,9,10]}
grid_search=GridSearchCV(rfr,param_grid)
```

```python
grid_result=grid_search.fit(X_train,y_train)
```

```python
grid_result.best_params_
```

```python
grid_result.score(X_train,y_train)
```

```python
grid_result.score(X_test,y_test)
```

```python
y_pred_rfr=grid_result.predict(X_test)
rfr_df=pd.DataFrame()
rfr_df['Target']=y_test
rfr_df['Predictions']=y_pred_rfr
mse_rfr=np.sqrt(mean_squared_error(y_pred_rfr,y_test))
print('Root mean squared error of Random Forest regression : {0:.3f}'.format(mse_rfr))
```

```python

sns.lmplot('Target','Predictions',data=rfr_df,line_kws={'color':'orange'},height=6,aspect=2)
plt.title('Random Forest regression \n RMSE:{0:.4f}'.format(mse_rfr),size=15)
```

Unexpectedly, Random Forest regressor actually performed worse than simple linear and ridge regression.

Let's check a gradient boosted algorithm such as XGBoost.

### XGBoost

```python
from xgboost import XGBRegressor
```

```python
xgb=XGBRegressor()
xgb.fit(X_train,y_train)
```

```python
xgb.score(X_train,y_train)
```

```python
xgb.score(X_test,y_test)
```

```python
y_pred_xgb=xgb.predict(X_test)
xgb_df=pd.DataFrame()
xgb_df['Target']=y_test
xgb_df['Predictions']=y_pred_xgb
mse_xgb=np.sqrt(mean_squared_error(y_pred_xgb,y_test))
print('Root mean squared error of XGB regression : {0:.4f}'.format(mse_xgb))
```

There is only a marginal improvement from Random Forest regression and still has underperformed when compared to Linear regression.

```python
sns.lmplot('Target','Predictions',data=xgb_df,line_kws={'color':'green'},height=6,aspect=2,scatter_kws={'alpha':0.5})
plt.title('XGBoosted regression \n RMSE:{0:.4f}'.format(mse_xgb),size=15)
```

It seems even the complex tree based algorithms couldn't bring down the error.

Let us check if LightGBM can give us better results.

Again, we see that LGBM has fared worse than the linear regression results.

We can safely say that given the data is normally distributed with no multicollinearity, Linear Regression maybe the best algorithm as was the case in our exercise here.
